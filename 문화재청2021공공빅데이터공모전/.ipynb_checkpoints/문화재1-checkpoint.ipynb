##### 각 목조문화재별 가장 가까운 소방서의 거리를 찾아서 칼럼 추가

```python
# 목조문화재 len = 478 , 소방서 len = 224
목조문화재['최근접소방서명'] = ""
목조문화재['최근접소방서거리'] = 1.0
for i in range(0,len(목조문화재)):
    min_len = 987654321
    name = ""
    for j in range(0,len(소방서)):
        first = (목조문화재.위도[i], 목조문화재.경도[i])
        second = (소방서.위도[j], 소방서.경도[j])
        temp = haversine(first,second, unit='km')
        if(min_len > temp):
            min_len = temp
            name = 소방서.소방서[j]
        if(j == len(소방서)-1):
            목조문화재.최근접소방서명[i] = name
            목조문화재.최근접소방서거리[i] = round(min_len,2)

목조문화재[['최근접소방서거리','최근접소방서명']]
```

#####  '수량/면적' 컬럼 숫자 데이터만 추출 및 결측치 처리

```python
목조문화재['수량/면적'].values
```

```python
목조문화재['수량/면적'] = 목조문화재['수량/면적'].replace(0, '0')  # temp = '㎡' in 목조문화재['수량/면적'][i] 적용 위해 변형

logic = []
for i in range(len(목조문화재)):
    temp = '㎡' in 목조문화재['수량/면적'][i]
    logic.append(temp)  
    
n_logic = []
for i in range(len(목조문화재)):
    temp = '㎡' not in 목조문화재['수량/면적'][i]
    n_logic.append(temp)  
    
print(목조문화재['수량/면적'][logic])
print(목조문화재['수량/면적'][n_logic])
```

```python
import re

for i in 목조문화재['수량/면적'][logic].index:
    string = 목조문화재['수량/면적'][i]
    value = re.findall(r'\b\d{1,3}(?:,\d{3})*(?:\.\d+)?(?!\d)', string)
    #print(value)
    
    for j in range(len(value)):
        value[j] = value[j].replace(',', '') # , 제거
        value[j] = float(value[j]) 
        #print(value)
            
    value = max(value) # 지정면적
    #print(value)
    
    목조문화재['수량/면적'][i] = value

목조문화재['수량/면적'][logic]
```

```python
# 결측치를 가지는 문화재의 다수가 1동 규모의 건조물
목조문화재[['문화재명','수량/면적']][n_logic].values 
```

```python
# 결측치를 1동 규모 건조물들의 평균값으로 대체
목조문화재['수량/면적'][n_logic] = 목조문화재['수량/면적'][logic][목조문화재['수량/면적'][logic] <= 100].mean()
목조문화재['수량/면적'][n_logic]
```

##### '최근접소방서거리', '수량/면적' 이상치 제거

```python
def get_outlier(data, column): 
    global lower, upper    
    
    q25 = np.quantile(data[column], 0.25)
    q75 = np.quantile(data[column], 0.75)          
    
    iqr = q75 - q25    
     
    upper = q75 + iqr * 1.5          
    lower = q25 - iqr * 1.5          
    
    data1 = data[data[column] > upper]     
    data2 = data[data[column] < lower]  
        
    print("상한 이상의 이상치 :",data1[['문화재명',column]].values)
    print()
    print("하한 이하의 이상치 :",data2[['문화재명',column]].values)
    print(iqr, q75, q25, upper, lower)
```

```python
get_outlier(목조문화재, '수량/면적')
```

```python
# '수량/면적' 이상치 제거
목조문화재_이상치제거 = 목조문화재[(lower <= 목조문화재['수량/면적']) & (목조문화재['수량/면적'] <= upper)]
```

```python
ks = range(1, 10)
sse = []

for k in ks:
    model=KMeans(n_clusters=k, algorithm = 'auto')
    model.fit(feature) #defalt : init='random', init=10, max_iter=100
    sse.append(model. inertia_)

sse
```

```python
plt.plot(ks, sse, '-o')
plt.ticklabel_format(style='plain')
plt.xlabel('number of clusters of k')
plt.ylabel('SSE')
plt.xticks(ks)
plt.show()
```

```python
scaler = StandardScaler()
model = KMeans(n_clusters=5, algorithm='auto', random_state=0)
pipeline = make_pipeline(scaler,model)
pipeline.fit(feature)
```
