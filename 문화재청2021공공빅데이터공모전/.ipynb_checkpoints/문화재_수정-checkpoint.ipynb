```python
import sys
import pandas as pd
import numpy as np
from haversine import haversine
import warnings

warnings.filterwarnings(action='ignore') 

# excel read
국보_보물 = pd.read_excel('국보,보물 2020.xlsx', header = [3,4,5,6]).fillna(0)
사적 = pd.read_excel('사적.xlsx', header = [4]).fillna(0)
국가민속문화재 = pd.read_excel('국가민속문화재.xlsx', header = [4]).fillna(0)
소방서 = pd.read_csv('소방청_시도 소방서 현황_20200121.csv', header=[0],encoding='CP949').fillna(0)

# 국보_보물 column 
col = []
for i in range(0,6):
    col.append(국보_보물.columns[i][0])        
for i in range(6,23):
    col.append((국보_보물.columns[i][3]))
국보_보물.columns = col

# concat 데이터 붙이기
사적_국가민속문화재 = pd.concat([사적, 국가민속문화재])
사적_국가민속문화재.drop(['Unnamed: 10', 'Unnamed: 11', 'Unnamed: 12'], axis=1, inplace=True)
# merge로 데이터 붙이기
목조문화재 = pd.merge(국보_보물, 사적_국가민속문화재, how='outer')

# 사적의 명칭이였던 칼럼을 문화재명으로 변경하고 명칭 칼럼 삭제
목조문화재.문화재명[len(국보_보물)+1:] = 목조문화재.명칭[len(국보_보물)+1:] # warning
목조문화재.drop('명칭', axis=1, inplace=True)

# 사적 위경도 split하여 위도 경도로 변경
for i in range(len(국보_보물),len(목조문화재)):
    if 목조문화재.위경도[i] != 0:
        temp = 목조문화재.위경도[i]
        temp = temp.split(", ")
        목조문화재.위도[i] = temp[0] # warning
        목조문화재.경도[i] = temp[1]

# 국보_보물의 시도,시군구 통합하여 소재지로 칼럼 변경
for i in range(0,len(국보_보물)):
    목조문화재.소재지[i] = '{} {}'.format(목조문화재.시도[i],목조문화재.시군구[i])    
목조문화재.drop(['시도', '시군구'], axis=1, inplace=True)

# 사용하지 않을 연번,위경도 칼럼 삭제
목조문화재.drop(['연번','위경도'], axis=1, inplace=True)
# \n이 포함된 칼럼명 수정
목조문화재.rename(columns = {'할론 \n또는 \n이산화탄소' : '할론 또는 이산화탄소'}, inplace = True)

# 목조문화재 len = 478 , 소방서 len = 224
# 각 목조문화재별 가까운 소방서의 거리를 찾아서 칼럼 추가
목조문화재['가까운소방서명'] = ""
목조문화재['가까운소방서거리'] = 1.0
for i in range(0,len(목조문화재)):
    min_len = 987654321
    name = ""
    for j in range(0,len(소방서)):
        first = (목조문화재.위도[i], 목조문화재.경도[i])
        second = (소방서.위도[j], 소방서.경도[j])
        temp = haversine(first,second)
        if(min_len > temp):
            min_len = temp
            name = 소방서.소방서[j]
    목조문화재.가까운소방서명[i] = name
    목조문화재.가까운소방서거리[i] = round(min_len,2)
print(목조문화재[['가까운소방서거리','가까운소방서명']])

목조문화재['가까운소방서거리'].min()
```
