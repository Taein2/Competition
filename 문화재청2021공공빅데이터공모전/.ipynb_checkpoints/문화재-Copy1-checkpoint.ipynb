```python
목조문화재.fillna(0, inplace=True)
소방서.fillna(0, inplace=True)
```

#####  '명칭' 컬럼의 데이터를 '문화재명'로 옮기고, '명칭' 컬럼 삭제

```python
목조문화재.문화재명[len(국보_보물):] = 목조문화재.명칭[len(국보_보물):]
목조문화재.drop('명칭', axis=1, inplace=True)

목조문화재.문화재명[len(국보_보물):] 
```

#####  '위경도' 컬럼 데이터를 '위도', '경도'로 분할

```python
# 사적 위경도 split하여 위도 경도로 변경
for i in range(len(국보_보물),len(목조문화재)):
    if 목조문화재.위경도[i] != 0:
        temp = 목조문화재.위경도[i]
        temp = temp.split(", ")
        목조문화재.위도[i] = temp[0] # warning
        목조문화재.경도[i] = temp[1]
            
목조문화재[['위도','경도']]
```

##### '시도' ,'시군구' 컬럼 데이터를 '소재지'로 통합하고, '시도' ,'시군구' 컬럼 삭제

```python
for i in range(0,len(국보_보물)):
    목조문화재.소재지[i] = '{} {}'.format(목조문화재.시도[i],목조문화재.시군구[i])
    
목조문화재.drop(['시도', '시군구'], axis=1, inplace=True)
```

```python
목조문화재.rename(columns = {'할론 \n또는 \n이산화탄소' : '할론 또는 이산화탄소'}, inplace = True)
```

##### 각 목조문화재별 가장 가까운 소방서의 거리를 찾아서 칼럼 추가

```python
# 목조문화재 len = 478 , 소방서 len = 224
목조문화재['최근접소방서명'] = ""
목조문화재['최근접소방서거리'] = 1.0
for i in range(0,len(목조문화재)):
    min_len = 987654321
    name = ""
    for j in range(0,len(소방서)):
        first = (목조문화재.위도[i], 목조문화재.경도[i])
        second = (소방서.위도[j], 소방서.경도[j])
        temp = haversine(first,second, unit='km')
        if(min_len > temp):
            min_len = temp
            name = 소방서.소방서[j]
        if(j == len(소방서)-1):
            목조문화재.최근접소방서명[i] = name
            목조문화재.최근접소방서거리[i] = round(min_len,2)

목조문화재[['최근접소방서거리','최근접소방서명']]
```

#####  '수량/면적' 컬럼 숫자 데이터만 추출 및 결측치 처리

```python
목조문화재['수량/면적'].values
```

```python
목조문화재['수량/면적'] = 목조문화재['수량/면적'].replace(0, '0')  # temp = '㎡' in 목조문화재['수량/면적'][i] 적용 위해 변형

logic = []
for i in range(len(목조문화재)):
    temp = '㎡' in 목조문화재['수량/면적'][i]
    logic.append(temp)  
    
n_logic = []
for i in range(len(목조문화재)):
    temp = '㎡' not in 목조문화재['수량/면적'][i]
    n_logic.append(temp)  
    
print(목조문화재['수량/면적'][logic])
print(목조문화재['수량/면적'][n_logic])
```

```python
import re

for i in 목조문화재['수량/면적'][logic].index:
    string = 목조문화재['수량/면적'][i]
    value = re.findall(r'\b\d{1,3}(?:,\d{3})*(?:\.\d+)?(?!\d)', string)
    # print(value)
    
    for j in range(len(value)):
        value[j] = value[j].replace(',', '') # , 제거
        value[j] = float(value[j]) 
        # print(value)
            
    value = max(value) # 지정면적
    # print(value)
    
    목조문화재['수량/면적'][i] = value

목조문화재['수량/면적'][logic]
```

```python
# 결측치를 가지는 문화재의 다수가 1동 규모의 건조물
목조문화재[['문화재명']][n_logic].values 
```

```python
# 결측치를 1동 규모 건조물들의 평균값으로 대체
목조문화재['수량/면적'][n_logic] = 목조문화재['수량/면적'][logic][목조문화재['수량/면적'][logic] <= 100].mean()
목조문화재['수량/면적'][n_logic]
```
